```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

# NLTK 불용어 다운로드 (한 번만 실행)
nltk.download('stopwords')

# 데이터 불러오기
data = pd.read_csv('news_summary_more.csv', encoding='iso-8859-1')

# 1. 결측값 제거
data.dropna(inplace=True)

# 2. 중복 데이터 제거
data.drop_duplicates(subset=['text', 'headlines'], inplace=True)

# 3. 텍스트 정규화 (소문자 변환, 특수문자 제거)
def clean_text(text):
    text = text.lower()  # 소문자로 변환
    text = re.sub(r'\s+', ' ', text)  # 여러 개의 공백을 하나로 변경
    text = re.sub(r"[^a-zA-Z0-9.,!?'\s]", '', text)  # 특수문자 제거
    return text.strip()

data['text'] = data['text'].apply(clean_text)
data['headlines'] = data['headlines'].apply(clean_text)

# 4. 불용어 제거 (선택적)
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

# headlines도 불용어 제거할지 고민 필요!
data['text'] = data['text'].apply(remove_stopwords)


# 5. 길이 제한 (너무 짧거나 긴 문장 제외)
def length_filter(text, min_len=10, max_len=500):
    return min_len <= len(text.split()) <= max_len

data = data[data['text'].apply(lambda x: length_filter(x, min_len=10, max_len=500))]
data = data[data['headlines'].apply(lambda x: length_filter(x, min_len=3, max_len=30))]

# 결과 확인
print(data.head())
print(f"전처리 후 데이터 개수: {len(data)}")
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense, Concatenate, Attention, TimeDistributed # TimeDistributed 추가!
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
import warnings
warnings.filterwarnings("ignore")

# Step 1 & 2에서 전처리된 데이터 로드 (가정)
# 실제로는 Step 2의 전처리 코드를 먼저 실행해야 합니다.
data = pd.read_csv('news_summary_more.csv', encoding='iso-8859-1')

# 전처리 함수 (Step 2 코드 재사용)
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r"[^a-zA-Z0-9.,!?'\s]", '', text)
    return text.strip()

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

nltk.download('stopwords') # 불용어 다운로드 (필요한 경우)

data.dropna(inplace=True)
data.drop_duplicates(subset=['text', 'headlines'], inplace=True)
data['text'] = data['text'].apply(clean_text)
data['headlines'] = data['headlines'].apply(clean_text)
data['text'] = data['text'].apply(remove_stopwords)
data = data[data['text'].apply(lambda x: 10 <= len(x.split()) <= 500)]
data = data[data['headlines'].apply(lambda x: 3 <= len(x.split()) <= 30)]


# 데이터 준비
encoder_input = data['text']
decoder_input = data['headlines']
decoder_target = data['headlines'].apply(lambda x : '<start> ' + x + ' <end>') # decoder target에 <start>, <end> 토큰 추가


# 학습 데이터, 검증 데이터 분리 (decoder_target 사용)
encoder_input_train, encoder_input_val, decoder_input_train, decoder_input_val, decoder_target_train, decoder_target_val = train_test_split(
    encoder_input, decoder_input, decoder_target, test_size=0.2, random_state=42
)

# 토크나이저 정의 및 학습
x_tokenizer = Tokenizer()
x_tokenizer.fit_on_texts(encoder_input_train)
encoder_vocab_size = len(x_tokenizer.word_index) + 1  # 단어 집합 크기

y_tokenizer = Tokenizer()
y_tokenizer.fit_on_texts(decoder_input_train)
decoder_vocab_size = len(y_tokenizer.word_index) + 1  # 요약 단어 집합 크기


# 정수 시퀀스로 변환
encoder_input_train = x_tokenizer.texts_to_sequences(encoder_input_train)
encoder_input_val = x_tokenizer.texts_to_sequences(encoder_input_val)
decoder_input_train = y_tokenizer.texts_to_sequences(decoder_input_train)
decoder_input_val = y_tokenizer.texts_to_sequences(decoder_input_val)
decoder_target_train = y_tokenizer.texts_to_sequences(decoder_target_train)
decoder_target_val = y_tokenizer.texts_to_sequences(decoder_target_val)


# 패딩 (최대 길이 맞추기) - decoder_target_train 기준으로 decoder_max_len 재계산
encoder_max_len = max(len(seq) for seq in encoder_input_train)
decoder_max_len = max(len(seq) for seq in decoder_target_train) # decoder_target_train 기준으로 변경!

print(f"Encoder Max Length: {encoder_max_len}") # Debugging print
print(f"Decoder Max Length: {decoder_max_len}")   # Debugging print

encoder_input_train = pad_sequences(encoder_input_train, maxlen=encoder_max_len, padding='post')
encoder_input_val = pad_sequences(encoder_input_val, maxlen=encoder_max_len, padding='post')
decoder_input_train = pad_sequences(decoder_input_train, maxlen=decoder_max_len, padding='post') # 새로운 decoder_max_len 적용
decoder_input_val = pad_sequences(decoder_input_val, maxlen=decoder_max_len, padding='post') # 새로운 decoder_max_len 적용
decoder_target_train = pad_sequences(decoder_target_train, maxlen=decoder_max_len, padding='post') # 새로운 decoder_max_len 적용
decoder_target_val = pad_sequences(decoder_target_val, maxlen=decoder_max_len, padding='post') # 새로운 decoder_max_len 적용

# Verify padded lengths - crucial debugging step
print(f"Max length of padded decoder_target_train (before slicing): {max(len(seq) for seq in decoder_target_train)}") # Debugging print
print(f"Max length of padded decoder_target_train (after slicing): {max(len(seq) for seq in decoder_target_train[:,1:])}") # Debugging print


# 모델 설계 (어텐션 메커니즘을 사용한 Seq2Seq)
embedding_dim = 128
hidden_dim = 256

# 1. 인코더
encoder_inputs = Input(shape=(encoder_max_len,))
enc_emb = Embedding(encoder_vocab_size, embedding_dim)(encoder_inputs)

# 인코더 LSTM 층 (stacked LSTM, 2 layers)
encoder_lstm1 = LSTM(hidden_dim, return_sequences=True, return_state=True, dropout=0.4, recurrent_dropout=0.4)
encoder_output1, state_h1, state_c1 = encoder_lstm1(enc_emb)

encoder_lstm2 = LSTM(hidden_dim, return_sequences=True, return_state=True, dropout=0.4, recurrent_dropout=0.4)
encoder_output2, state_h2, state_c2 = encoder_lstm2(encoder_output1)

# 2. 디코더
decoder_inputs = Input(shape=(decoder_max_len,)) # decoder_max_len으로 수정!
dec_emb_layer = Embedding(decoder_vocab_size, embedding_dim)
dec_emb = dec_emb_layer(decoder_inputs)

# 디코더 LSTM 층 (stacked LSTM, 2 layers)
decoder_lstm1 = LSTM(hidden_dim, return_sequences=True, return_state=True, dropout=0.4, recurrent_dropout=0.2)
decoder_output1, decoder_state_h1, decoder_state_c1 = decoder_lstm1(dec_emb, initial_state=[state_h1, state_c1])

decoder_lstm2 = LSTM(hidden_dim, return_sequences=True, return_state=True, dropout=0.4, recurrent_dropout=0.2)
decoder_output2, decoder_state_h2, decoder_state_c2 = decoder_lstm2(decoder_output1, initial_state=[state_h2, state_c2])


# 3. 어텐션 층 (Bahdanau 스타일)
attn_layer = Attention()
attn_output = attn_layer([decoder_output2, encoder_output2])

# 4. 어텐션 결과와 디코더 hidden state 연결 및 출력층
decoder_concat_input = Concatenate(axis=-1, name='concat_layer')([decoder_output2, attn_output])

decoder_dense = TimeDistributed(Dense(decoder_vocab_size, activation='softmax')) # TimeDistributed 적용!
decoder_outputs = decoder_dense(decoder_concat_input)


# 모델 정의
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# 학습
# 설계한 모델로 훈련 진행
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy')
es = EarlyStopping(monitor='val_loss', patience=2, verbose=1)
history = model.fit(x=[encoder_input_train, decoder_input_train], y=decoder_target_train, \
          validation_data=([encoder_input_test, decoder_input_test], decoder_target_test), \
          batch_size=256, callbacks=[es], epochs=50)


print("Step 3. 어텐션 메커니즘을 사용한 Seq2Seq 모델 학습 완료!")
```
